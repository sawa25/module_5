```python
import pandas as pd
import numpy as np
from pandas import Series


import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_selection import f_classif, mutual_info_classif
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV


from sklearn.metrics import confusion_matrix
from sklearn.metrics import auc, roc_auc_score, roc_curve,precision_recall_curve,average_precision_score
from sklearn.metrics import log_loss,accuracy_score,precision_score, recall_score,f1_score
```

```python
train=vis_data
num_cols=['age','decline_app_cnt','income','bki_request_cnt']
for i in num_cols:
    train[i] = train[i].apply(lambda w: np.log(w + 1))
for i in num_cols:
    plt.figure()
    sns.displot(train[i][train[i] > 0].dropna(), kde = False, rug=False)
    plt.title(i)
    plt.show()
```

```python
for comb in [('age','decline_app_cnt'),('income','bki_request_cnt')]:
    fig, ax = plt.subplots(1, 2, squeeze=False, figsize=(15, 6))
    #sns.jointplot(data=reduced_kids, x=comb[0], y=comb[1],ax=ax[count]) #
    ax[0, 0].set_title(comb[0])
    ax[0, 1].set_title(comb[1])
    
    sns.boxplot(x="default", y=comb[0], data=train,ax=ax[0, 0])
    sns.boxplot(x="default", y=comb[1], data=train,ax=ax[0, 1])
#         axes = sns.barplot(data=vis_data,
#                            x=comb[0], y=comb[1], hue='score', ax=ax[0, 0])
#     axes.legend(loc=4, bbox_to_anchor=(0.1, 0, 1, 1))
    plt.tight_layout()
```

```python
sns.heatmap(train[num_cols].corr().abs(), vmin=0, vmax=1)
```

```python
imp_num = pd.Series(f_classif(train[num_cols], train['default'])[0], index = num_cols)
imp_num.sort_values(inplace = True)
imp_num.plot(kind = 'barh')
```

```python
train.head()
```

```python
# Для бинарных признаков мы будем использовать LabelEncoder

label_encoder = LabelEncoder()

for column in bin_cols:
    train[column] = label_encoder.fit_transform(train[column])
    
# убедимся в преобразовании    
train.head()
```

```python
train['education'] = label_encoder.fit_transform(train['education'])
    
# убедимся в преобразовании    
train.head()
```

```python
imp_cat = Series(mutual_info_classif(train[bin_cols + cat_cols], train['default'],
                                     discrete_features =True), index = bin_cols + cat_cols)
imp_cat.sort_values(inplace = True)
imp_cat.plot(kind = 'barh')
```

```python
X_cat = OneHotEncoder(sparse = False).fit_transform(train[cat_cols].values)
X_cat.shape
```

```python
# Стандартизация числовых переменных

X_num = StandardScaler().fit_transform(train[num_cols].values)
X_num
```

```python
# Объединяем

X = np.hstack([X_num, train[bin_cols].values, X_cat])
Y = train['default'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

```python
def pltroc(fpr, tpr,roc_auc):
    plt.figure()
    plt.plot([0, 1], label='Baseline', linestyle='--')
    plt.plot(fpr, tpr, label = 'Regression')
    plt.title('Logistic Regression ROC AUC = %0.3f' % roc_auc)
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend(loc = 'lower right')
    plt.show()

model = LogisticRegression(max_iter=300)
model.fit(X_train, y_train)
probs = model.predict_proba(X_test)
probs = probs[:,1]
fpr, tpr, threshold = roc_curve(y_test, probs)
roc_auc = roc_auc_score(y_test, probs)
pltroc(fpr, tpr,roc_auc)
```

```python
from sklearn.model_selection import GridSearchCV

# Добавим типы регуляризации
penalty = ['l1', 'l2']

# Зададим ограничения для параметра регуляризации
C = np.logspace(0, 4, 10)

# Создадим гиперпараметры
hyperparameters = dict(C=C, penalty=penalty)

model = LogisticRegression(max_iter=300)
model.fit(X_train, y_train)

# Создаем сетку поиска с использованием 5-кратной перекрестной проверки
clf = GridSearchCV(model, hyperparameters, cv=5, verbose=0)

best_model = clf.fit(X_train, y_train)

# View best hyperparameters
print('Лучшее Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Лучшее C:', best_model.best_estimator_.get_params()['C'])
```

```python
best_model.fit(X_train, y_train)
probs = best_model.predict_proba(X_test)
probs = probs[:,1]
fpr, tpr, threshold = roc_curve(y_test, probs)
roc_auc = roc_auc_score(y_test, probs)
pltroc(fpr, tpr,roc_auc)
```

```python
model = LogisticRegressionCV(solver='liblinear') #max_iter=300
model.fit(X_train, y_train)

# Добавим типы регуляризации
penalty = ['l1','l2']

# Зададим ограничения для параметра регуляризации
C =[1,2,3,4,5]  #np.logspace(0, 4, 10)

# Создадим гиперпараметры
hyperparameters = dict(Cs=C, penalty=penalty)

# Создаем сетку поиска с использованием 5-кратной перекрестной проверки
clf = GridSearchCV(model, hyperparameters, cv=5, verbose=0)

best_model = clf.fit(X_train, y_train)

```

```python
# View best hyperparameters
print('Лучшее Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Лучшее C:', best_model.best_estimator_.get_params()['Cs'])
probs = model.predict_proba(X_test)
y_pred= model.predict(X_test)
probs = probs[:,1]
fpr, tpr, threshold = roc_curve(y_test, probs)
roc_auc = roc_auc_score(y_test, probs)
pltroc(fpr, tpr,roc_auc)
```

```python
#вычисление log_loss
y_hat = np.zeros((len(y_test), 2), dtype=np.float32)
for num in range(2):
    y_hat[:, num] = 1. * (y_pred == num)
print('log_loss: {:.4f}'.format(log_loss(y_test, y_hat)))
acc = accuracy_score(y_test, y_pred)
print('Accuracy from scikit-learn: {}%'.format(acc * 100))
precision = precision_score(y_test, 1 - y_pred)
recall = recall_score(y_test, 1 - y_pred)
print('precision: {:.2f}\nrecall: {:.2f}'.format(precision, recall))
f1 = f1_score(y_test, y_pred)
print('F1-score: {:.2f}'.format(f1))
ap = average_precision_score(y_test, y_pred)
print('Average Precision (AP) = {:.2f}'.format(ap))
```
