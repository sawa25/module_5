```python
bin_cols = []
cat_cols = []
num_cols = []
# разбивка столбцов датасета на бинарные, категориальные и числовые признаки по спискам
def splitcolumnsbyt(vis_data,postperform=None,verb=None):
    global bin_cols
    global cat_cols
    global num_cols
    bin_cols = []
    cat_cols = []
    num_cols = []
    if not verb is None:   print("разбивка столбцов на бинарные, категориальные и числовые признаки")
    for i in vis_data.columns:
        tmp = vis_data[i].unique()
        if len(tmp) == 2:
            if i != 'default':
                bin_cols.append(i)
        elif len(tmp) < 10:
            cat_cols.append(i)
        else:
            if i != 'client_id':
                num_cols.append(i)
                
    if postperform is None:
        pass
    else: postperform()

    if not verb is None:
        display("бинарные:", bin_cols, "категориальные:",
            cat_cols, "числовые:", num_cols)
        print("отображение многообразия вариаций нечисловых столбцов")
        for i in (bin_cols+cat_cols):
            print(i, ":", vis_data[i].unique())


splitcolumnsbyt(vis_data,verb=True)
```

```python
# рассмотреть распределение числовых признаков
def numplothist(df,diagtype="hist"):
    for i in range(0, len(num_cols), 2):
        fig, ax = plt.subplots(1, 2, squeeze=False, figsize=(25, 10))
        for axi in range(2):
            for labelx, labely in zip(ax[0, axi].xaxis.get_ticklabels(), ax[0, axi].yaxis.get_ticklabels()):
                labelx.set_fontsize(20)
                labely.set_fontsize(20)
        comb = [num_cols[i]]
        ax[0, 0].set_title(comb[0], fontsize='xx-large')
        if i < len(num_cols)-1:
            comb.append(num_cols[i+1])
            ax[0, 1].set_title(comb[1], fontsize='xx-large')
            if diagtype == "boxplot":
                ax[0, 0].set_xlabel('default',fontsize='xx-large')
                ax[0, 1].set_xlabel('default',fontsize='xx-large')
                sns.boxplot(x="default", y=comb[0], data=df, ax=ax[0, 0])
                sns.boxplot(x="default", y=comb[1], data=df, ax=ax[0, 1])
            else:
                df[comb[0]].hist(ax=ax[0, 0], bins=100)
                df[comb[1]].hist(ax=ax[0, 1], bins=100)
        else:
            if diagtype == "boxplot":
                ax[0, 0].set_xlabel('default',fontsize='xx-large')
                sns.boxplot(x="default", y=comb[0], data=df, ax=ax[0, 0])
            else:
                df[comb[0]].hist(ax=ax[0, 0], bins=100)
        plt.tight_layout()


numplothist(vis_data)
```

```python
#визуализация соотношения вариантов целевой переменной в тренировочной и тестовой выборках
def visualize_train_valid_counts(init_data, y_train, y_test):
    x = np.array([0, 1])
    width = 0.4

    plt.figure(figsize=(15, 8))
    ax = plt.subplot(111)

    classes = list(init_data['default'].value_counts().index)

    ax.bar(x - width, list(init_data['default'].value_counts()[classes]), width, color='r', label='Исходные данные')
    ax.bar(x, [np.sum(y_train==0),np.sum(y_train==1)], width, color='g', label='Тренировочная выборка')
    ax.bar(x, [np.sum(y_test==0),np.sum(y_test==1)], width, bottom=[np.sum(y_train==0),np.sum(y_train==1)], color='b', label='Валидационная выборка')

#     ax.set_ylim([0, 70])
    plt.xticks(x - width / 2, classes, fontsize=20)
    plt.yticks(fontsize=15)
    plt.xlabel('Значение целевой переменной default', fontsize=20)
    plt.ylabel('Кол-во примеров', fontsize=20)
    plt.minorticks_on()
    plt.grid(which='major', color='r')
    plt.grid(which='minor', linestyle=':', color='k')
    plt.legend(fontsize=15)
```

```python
def pltroc(fpr, tpr, roc_auc):  # построение графика ROC AUC
    plt.figure()
    plt.plot([0, 1], label='Baseline', linestyle='--')
    plt.plot(fpr, tpr, label='Regression')
    plt.title('Logistic Regression ROC AUC = %0.3f' % roc_auc)
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend(loc='lower right')
    plt.show()
```

```python
class LogReg():
    def __init__(self, df,bin_cols,cat_cols,num_cols):
        self.df=df
        #кодирование категориальных переменных
        self.X_cat = OneHotEncoder(sparse = False).fit_transform(df[cat_cols].values)
        # Стандартизация числовых переменных
        self.X_num = StandardScaler().fit_transform(df[num_cols].values)
        # Объединение подготовленных значений в единую группу для передачи в модель
        self.X = np.hstack([self.X_num, df[bin_cols].values, self.X_cat])
        self.Y = df['default'].values
        self.probs,self.model,self.X_train, self.X_test, self.y_train, self.y_test=None,None,None,None,None,None
        self.clf,self.C = None,None
        print("Создан объект класса LogReg")
    
    def traintest_split(self):
        #разделение выборки на тренировочную и тестовую
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, self.Y, test_size=0.20, random_state=42)        
    
    def pltsplit(self):
        if self.y_train is None:
            raise Exception('Call traintest_split first')
        visualize_train_valid_counts(self.df, self.y_train, self.y_test)
    
    def pltrocauc(self):
        if self.probs is None:
            raise Exception('Call predict first')
        fpr, tpr, threshold = roc_curve(self.y_test, self.probs[:,1])
        roc_auc = roc_auc_score(self.y_test, self.probs[:,1])
        pltroc(fpr, tpr, roc_auc)
        
    def fit(self):
        if self.y_train is None:
            raise Exception('Call traintest_split first')
        self.model = LogisticRegression(max_iter=300)
        self.model.fit(self.X_train, self.y_train)
        
    def searchCV(self,C):
        self.C=C #np.logspace(0, 4, num=5)
        hyperparameters = dict(C=C) #, penalty=penalty
        #self.model.fit(self.X_train, self.y_train)

        # Создаем сетку поиска с использованием 5-кратной перекрестной проверки
        self.clf = GridSearchCV(self.model, hyperparameters, cv=5, verbose=3)
        print('Start search C')
        self.best_model = self.clf.fit(self.X_train, self.y_train)
        print('End search C')
        print('Лучшее C:', self.best_model.best_estimator_.get_params()['C'])
        self.model=self.best_model.best_estimator_

    def searchHalv(self,C,max_iterint):
        self.C=C
        hyperparameters = dict(C=C,max_iter=max_iterint) #, penalty=penalty

        # Создаем сетку поиска с использованием 5-кратной перекрестной проверки
        self.clf = HalvingGridSearchCV(self.model, hyperparameters, cv=5, verbose=3,factor=2, random_state=0)
        
        print('Start search C')
        self.best_model = self.clf.fit(self.X_train, self.y_train)
        print('End search C')
        print('Лучшее C:', self.best_model.best_estimator_.get_params()['C'])
        self.model=self.best_model.best_estimator_
        
    def predict(self):
        if self.model is None:
            raise Exception('Call fit first')
        self.probs = self.model.predict_proba(self.X_test)
        self.y_pred= self.model.predict(self.X_test)
        
    def printmetrics(self):
        if self.y_pred is None:
            raise Exception('Call predict first')
        #вычисление log_loss
        y_hat = np.zeros((len(self.y_test), 2), dtype=np.float32)
        for num in range(2):
            y_hat[:, num] = 1. * (self.y_pred == num)
        print('log_loss: {:.4f}'.format(log_loss(self.y_test, y_hat)))
        acc = accuracy_score(self.y_test, self.y_pred)
        print('Accuracy from scikit-learn: {}%'.format(acc * 100))
        precision = precision_score(self.y_test, 1 - self.y_pred)
        recall = recall_score(self.y_test, 1 - self.y_pred)
        print('precision: {:.2f}\nrecall: {:.2f}'.format(precision, recall))
        f1 = f1_score(self.y_test, self.y_pred)
        print('F1-score: {:.2f}'.format(f1))
        ap = average_precision_score(self.y_test, self.y_pred)
        print('Average Precision (AP) = {:.2f}'.format(ap))
        
    def pltprecrec(self):
        precisions, recalls, _ = precision_recall_curve(self.y_test, self.y_pred)
        plt.step(recalls, precisions, color='b', alpha=0.2, where='post')
        plt.fill_between(recalls, precisions, step='post', alpha=0.2, color='b')
        plt.xlabel('Recall');
        plt.ylabel('Precision');     
        
    def printconfmatrx(self):
        conf_mat = confusion_matrix(self.y_test, self.y_pred).T
        print('Confusion matrix:\n{}'.format(conf_mat))        
```
